## 1.1 What is Netflix <a id=1.1></a>

Netflix is a global streaming service that offers a wide variety of television shows, movies, documentaries, and other entertainment content across a broad range of genres and languages. It was founded in 1997 by Reed Hastings and Marc Randolph in Scotts Valley, California. Initially starting as a DVD rental service by mail, Netflix evolved into a subscription-based streaming platform in 2007, allowing members to instantly watch content on their internet-connected devices.

One of the defining features of Netflix is its vast library of content, which includes both licensed and original productions. Netflix Originals are exclusive titles produced or commissioned by Netflix itself, spanning various formats such as TV series, films, documentaries, and stand-up specials. These original productions have garnered critical acclaim and popularity, contributing significantly to Netflix's success as a leading streaming service.

Netflix operates on a subscription-based model, offering different subscription tiers with varying features, including video quality and simultaneous streams. Subscribers have access to a personalized viewing experience, with recommendations and curated content based on their viewing history and preferences.

The streaming service is available in over 190 countries worldwide, making it accessible to a diverse and global audience. Netflix is accessible on various devices, including smart TVs, smartphones, tablets, gaming consoles, and streaming media players, allowing users to enjoy content anytime, anywhere, with an internet connection.

In recent years, Netflix has become synonymous with binge-watching culture, where viewers consume multiple episodes or entire seasons of a series in a single sitting. Its innovative approach to content distribution, emphasis on user experience, and investment in original programming have made Netflix a dominant force in the entertainment industry, shaping the way people consume and interact with media.

## 4.3 Filling Missing Values Of `Director`, `Country`, `Cast` Variables

_It doesnt make sense to fill out missing values of these columns using statistical measurement, because their dtype is `Object` so it is not possible to fill them using mean, median._
_Also using mode is not good approach to fill them as mode will use most frequent value to fill NaNs and it will causes the accuracy and precision of the data to decrease and not be correct_
_So i will add `Unknown` value for these missing values to show that value of them are not known_

```python
df[['Director', 'Cast', 'Country']] = df[['Director', 'Cast', 'Country']].fillna('Unknown')
```

## 5.3 Search Top Directors Based on Most Movies They Have Produced.

```python
def top_directors(df, n=6):
  """
  Description:
      This function takes a DataFrame containing a 'Director' column and returns the top directors
      based on the frequency of their occurrence in the dataset. By default, it returns the top 6
      directors, but the user can specify the number of directors to be returned using the 'n' parameter.

  Parameters:
      - df (DataFrame): The input DataFrame containing a 'Director' column.
      - n (int, optional): The number of top directors to be returned. Default is 5.

  Returns:
      Series: A Pandas Series containing the top directors and their respective counts.

  Example:
      top_directors(df, n=10)

  """
  return df['Director'].value_counts().head(n).to_frame()
```

## 5.4 Number Of Movies/Series Released Per Year

```python
def release_year_analysis(df):
    """
    Description:
      This function performs analysis on the release years of content in a DataFrame. It extracts the
      release year from the 'Release_Date' column, calculates the frequency of content released in each
      year, and sorts the results by the release year.

    Parameters:
      - df (DataFrame): The input DataFrame containing a 'Release_Date' column with datetime values.

    Returns:
      Series: A Pandas Series containing the count of content released in each year, sorted by year.

    Example:
      release_year_analysis(df)

    """

    df['Release_Year'] = df['Release_Date'].dt.year
    return df['Release_Year'].value_counts().sort_index().to_frame()
```

```python
release_year_analysis(df)
```

## 5.5 Number Of Movies/Series Of Casts (Actors) <a id=5.5></a>

```python
def cast_analysis(df):
    """
    Description:
    This function performs analysis on the cast members of content in a DataFrame. It extracts
    individual cast members from the 'Cast' column, separates names and surnames if present, and
    counts the occurrences of each cast member. The function then returns a DataFrame containing
    the count of appearances for each cast member, sorted by the number of appearances in descending order.

    Parameters:
        - df (DataFrame): The input DataFrame containing a 'Cast' column with cast information.

    Returns:
        DataFrame: A Pandas DataFrame containing the count of appearances for each cast member, sorted by count.

    Example:
        cast_analysis(df)
    """
    # Create a DataFrame with individual cast members
    cast_df = pd.DataFrame(df['Cast'].apply(lambda x: str(x).split(',')).explode().str.strip())
    cast_df[['Name', 'Family']] = cast_df['Cast'].str.split(n=1, expand=True)     # Split name and family into separate columns
    cast_df = cast_df.drop_duplicates()          # Drop duplicates

    # Merge cast_df with the original DataFrame 'df' to get the Type (Movie/TV Show) for each cast member
    cast_df = cast_df.merge(df[['Cast']], left_on='Cast', right_on='Cast', how='left')

    # Group by name, family, count occurrences, and reset index
    cast_counts = cast_df.groupby(['Name', 'Family']).size().reset_index(name='Count').set_index(['Name', 'Family'])

    return cast_counts.sort_values(by = 'Count', ascending = False)
```

## 5.6 Number Of Movies/Series Of Directors <a id =5.6></a>

_This function counts the total number of movies and TV shows produced by each individual director by disregarding those records where director is Unknown_

```python
def director_production_count(df):
    """
    Function: director_production_count

    Description:
        This function calculates the production count of movies and TV shows for each director in a DataFrame.
        It filters out entries with 'Unknown' directors, extracts individual directors from the 'Director' column,
        counts the number of movies and TV shows produced by each director, and returns a DataFrame containing
        the production count for each director, sorted by count in descending order.

    Parameters:
        - df (DataFrame): The input DataFrame containing a 'Director' column with director information.

    Returns:
        DataFrame: A Pandas DataFrame containing the production count of movies and TV shows for each director,
                  sorted by count in descending order.

    Example:
        director_production_count(df)

    """
    # Filter out entries with 'Unknown' directors
    filtered_df = df[df['Director'] != 'Unknown']
    # Create a DataFrame with each unique directors
    director_df = pd.DataFrame(filtered_df['Director'].apply(lambda x: str(x).split(',')).explode().str.strip())
    # Count the total number of movie and TV show produced by each director
    director_production_counts = director_df.groupby('Director').size().reset_index(name='Count').set_index('Director')

    return director_production_counts.sort_values(by = 'Count', ascending = False)
```

```python
director_production_count(df).head(10)
```

```python
# Creating DataFrame for plotting
top_directors_df = director_production_count(df).head(10)
```

## 5.7 Number Of Movies Per Category <a id=5.7></a>

```python
def type_count(df):
    """Function: type_count

Description:
    This function calculates the count of content types (e.g., movie, TV show) in a DataFrame. It extracts
    individual types from the 'Type' column, removes duplicates, and counts the occurrences of each type.
    The function then returns a DataFrame containing the count of each content type, sorted by count in descending order.

Parameters:
    - df (DataFrame): The input DataFrame containing a 'Type' column with content type information.

Returns:
    DataFrame: A Pandas DataFrame containing the count of each content type, sorted by count in descending order.

Example:
    type_count(df)
"""
    # Creating new df by spliting genres in Type column
    type_df = pd.DataFrame(df['Type'].apply(lambda x: str(x).split(',')).explode().str.strip())
    type_df = type_df.drop_duplicates() # to drop duplicate genres
    type_df = type_df.merge(df['Type'], left_on = 'Type', right_on='Type', how = 'left') # merge
    type_count = type_df.groupby('Type').size().reset_index(name = 'Count').set_index('Type')
    return type_count.sort_values(by = 'Count', ascending = False)
```

```python
type_count(df)
```

```python
# Creating Dataframe for Plotting
top_genres_df = type_count(df).head(10)
```
