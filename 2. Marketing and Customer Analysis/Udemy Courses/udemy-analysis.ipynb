# 1. Introduction <a id=1></a>

## 1.1 About Udemy <a id=1.1></a>

Udemy is an online learning platform that offers a wide range of courses on various subjects, including technology, business, design, personal development, and more. It was founded in 2010 by Eren Bali, Gagan Biyani, and Oktay Caglar and has since grown to become one of the largest and most popular online course providers globally.

Here are some key features and aspects of Udemy:

1. **Course Variety**: Udemy hosts thousands of courses on diverse topics, ranging from programming languages like Python, JavaScript, and Java, to business skills like marketing, finance, and entrepreneurship. There are also courses on personal development, lifestyle, photography, music, and many other areas of interest.

2. **Instructor-Led Learning**: Courses on Udemy are created and taught by instructors from around the world who are experts in their respective fields. They provide video lectures, presentations, quizzes, assignments, and other learning materials to help students grasp the concepts effectively.

3. **Flexible Learning**: One of the key benefits of Udemy is its flexibility. Students can access courses anytime, anywhere, and learn at their own pace. There are no rigid schedules or deadlines, allowing learners to fit their studies around their busy lives.

4. **Lifetime Access**: Once enrolled in a course, students typically have lifetime access to the course materials, including updates and new content added by the instructor. This enables learners to revisit the material whenever they need to refresh their knowledge or learn new topics.

5. **Affordability**: Udemy courses are often priced affordably, especially when compared to traditional education options like college or university courses. Additionally, Udemy frequently offers discounts and promotions, making the courses even more accessible to a wide range of learners.

6. **User Reviews and Ratings**: Each course on Udemy has user reviews and ratings, allowing prospective students to gauge the quality and effectiveness of the course before enrolling. This helps learners make informed decisions about which courses to take.

7. **Certificates of Completion**: Upon completing a course on Udemy, students typically receive a certificate of completion. While these certificates may not carry the same weight as traditional degrees or certifications, they can still be valuable for showcasing new skills to employers or clients.

8. **Mobile App**: Udemy offers a mobile app for iOS and Android devices, allowing students to learn on the go. The app provides full access to course materials, enabling learners to watch lectures, complete assignments, and engage with the course community from their smartphones or tablets.

Overall, Udemy provides a convenient and accessible platform for lifelong learning, empowering individuals to acquire new skills, advance their careers, and pursue their passions at their own pace and on their own terms.

## 1.2 How Learning Platform Works <a id=1.2></a>

Online learning platforms like Udemy operate on a model that facilitates the creation, distribution, and consumption of educational content in a digital format. Here's an overview of how these platforms typically work:

1. **Course Creation**:
   - Instructors, subject matter experts, or educational organizations create courses on various topics based on their expertise and knowledge. Course content can include video lectures, presentations, quizzes, assignments, downloadable resources, and discussion forums.

2. **Course Submission**:
   - Instructors submit their courses to the online learning platform for review and approval. The platform's review team assesses the quality, relevance, and adherence to guidelines before accepting the course for publication.

3. **Course Publication**:
   - Once approved, the course is published on the platform's website or app, making it accessible to learners worldwide. Courses are typically categorized by subject, level of difficulty, duration, and other relevant criteria to help students find courses that match their interests and learning goals.

4. **Enrollment**:
   - Students browse the platform's catalog of courses and enroll in the ones that interest them. Enrollment usually involves purchasing the course at a one-time fee or subscribing to a membership plan that grants access to multiple courses.

5. **Learning Experience**:
   - Students access course materials through the platform's website or app. They progress through the course at their own pace, watching video lectures, completing assignments, taking quizzes, and participating in discussions with the instructor and other students.

6. **Support and Engagement**:
   - Online learning platforms often provide support to both instructors and students. Instructors may receive guidance on course creation, marketing, and engagement strategies. Students can reach out to instructors or platform support staff for assistance with technical issues, course content clarification, or general inquiries.

7. **Assessment and Certification**:
   - Courses may include quizzes, exams, or projects to assess students' understanding and mastery of the subject matter. Upon completing a course, students may receive a certificate of completion or achievement, which they can add to their resume or LinkedIn profile to showcase their newly acquired skills.

8. **Community and Networking**:
   - Many online learning platforms foster a sense of community among learners by providing discussion forums, social media groups, and networking opportunities. Students can connect with peers, share experiences, collaborate on projects, and seek advice from others in the community.

9. **Continuous Improvement**:
   - Online learning platforms continually strive to enhance the learning experience for both instructors and students. They gather feedback from users, analyze data on course performance, and iterate on their platform features and offerings to meet the evolving needs of learners.

Overall, online learning platforms like Udemy provide a flexible and accessible way for individuals to acquire new knowledge, develop skills, and advance their careers in a rapidly changing world. They democratize education by removing barriers to access and empowering learners to pursue lifelong learning opportunities from anywhere, at any time.

_`content_duration` Variable_

```python
def extract_numeric_value(value):
    """
    Extracts numeric values from a string representing duration in hours and/or minutes.

    Parameters:
    value (str): A string representing duration, e.g., '1.5 hours', '30 mins', '2 hours 45 mins'.

    Returns:
    int or float or None: The numeric value extracted from the input string. If the input string 
    contains 'hour' or 'hours', the value is converted to minutes and returned as an integer. 
    If the input string contains 'min' or 'mins', the value is returned as an integer. 
    If the input string does not contain any relevant information, None is returned.

    Example:
    >>> extract_numeric_value('1.5 hours')
    90
    >>> extract_numeric_value('30 mins')
    30
    >>> extract_numeric_value('2 hours 45 mins')
    165
    >>> extract_numeric_value('invalid string')
    None
    """
    try:
        if 'hour' in value or 'hours' in value:
            # Extract numeric value, convert to numeric type, and multiply by 60 to convert hours to minutes
            return pd.to_numeric(value.split()[0]) * 60
        elif 'min' in value or 'mins' in value:
            # Extract numeric value and convert to numeric type
            return pd.to_numeric(value.split()[0])
        else:
            return None  
    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
df['content_duration'] = df['content_duration'].apply(extract_numeric_value)
```

## 5.6 Number Of Courses Based on Subject & Level <a id=5.6></a>

```python
level_per_subj = df.groupby(['subject'])['level'].value_counts().to_frame()
level_per_subj
```

```python
# Reset index to make 'subject' and 'level' as columns
df_reset = level_per_subj.reset_index()

plt.figure(figsize=(8,6))
sns.barplot(x='count', y='subject', hue='level', data=df_reset, palette='rainbow')
plt.title('Count of Courses by Subject and Level', fontsize=12)
plt.xlabel('Count', fontsize=15)
plt.ylabel('Subject', fontsize=15)
plt.legend(title='Level', title_fontsize='12')
plt.show()
```

## 5.10 Histogram Charts <a id=5.10></a>

```python
def hist_chart(col):
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=col, kde=True, color='darkblue', bins='auto')
    plt.title('Distribution of Course Durations')
    plt.xlabel('Duration (minutes)')
    plt.ylabel('Frequency')
    plt.show()
```
