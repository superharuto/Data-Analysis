## 1.1 What is Data Science <a id=1.1></a>

Data science is an interdisciplinary field that deals with extracting insights and knowledge from data, both structured and unstructured, using scientific methods, algorithms, and systems. Here are some key points about data science:

1. **Interdisciplinary Nature**: Data science combines techniques and concepts from various disciplines such as statistics, mathematics, computer science, domain expertise, and subject matter knowledge.

2. **Data-driven Decision Making**: Data science enables organizations to make informed decisions based on data analysis and insights rather than relying solely on intuition or experience.

3. **Data Lifecycle**: The data science process typically involves several stages, including data acquisition, data cleaning, exploratory data analysis (EDA), feature engineering, modeling, evaluation, and deployment.

4. **Skills Required**: Data scientists need a combination of technical skills (programming, statistics, machine learning, data visualization) and domain knowledge to be effective in their roles.

5. **Tools and Technologies**: Data scientists use a variety of tools and technologies to work with data, such as programming languages (Python, R), data manipulation libraries (Pandas, NumPy), machine learning frameworks (Scikit-learn, TensorFlow, PyTorch), and visualization tools (Matplotlib, Seaborn, Tableau).

6. **Applications**: Data science is applied in numerous fields and industries, including healthcare, finance, retail, marketing, cybersecurity, social media analysis, climate science, and more. It helps organizations optimize processes, improve customer experiences, detect anomalies, predict trends, and innovate products and services.

7. **Ethical Considerations**: Data science raises important ethical considerations related to privacy, fairness, transparency, and accountability. Data scientists must adhere to ethical guidelines and regulations to ensure responsible use of data.

8. **Continuous Learning**: Data science is a rapidly evolving field, with new techniques, algorithms, and technologies emerging frequently. Continuous learning and staying updated with the latest developments are essential for data scientists to remain effectives various sectors.

## 4.2 Resolving Inconsistencies and Nosies <a id=4.2></a>

```python
def correctMinusOne(df, cols):
    """
        Corrects "-1" values in specified columns of the DataFrame by replacing them with NaN.

        Parameters:
            df (DataFrame): The pandas DataFrame containing the data.
            cols (list): A list of column names where "-1" values need to be corrected.
        
        Returns:
            None. The function modifies the DataFrame in place.
        
        Example:
            correctMinusOne(data, ['Size', 'Revenue'])
    """
    
    for col in cols:
        df.loc[(df[col] == '-1'), col] = np.nan
```

```python
# Call the Function
correctMinusOne(df, ['Headquarters', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'Founded'])
```

## 4.3 Extract Mix Salary/Max Salary from Salary Estimate Variable <a id=4.3></a>

```python
import re
def extractSalary(salary_string):
    """
        Extracts the minimum and maximum salary values from a salary string.

        Parameters:
        - salary_string (str): The salary string containing the salary range in the format "$minK-$maxK".
        
        Returns:
        - tuple: A tuple containing the minimum and maximum salary values extracted from the salary string. If no match is found, returns (None, None).

    """
    pattern = r'\$(\d+)K-\$(\d+)K'
    match = re.search(pattern, salary_string)
    if match:
        min_salary = int(match.group(1))
        max_salary = int(match.group(2))
        return min_salary, max_salary
    else:
        return None, None
```

```python
# Applying Method
df[["Min_Salary", "Max_Salary"]] = df['Salary Estimate'].apply(extractSalary).apply(pd.Series)
```

```python
def addThreeZeros(df, cols):
    for col in cols:
        df[col] = df[col].astype(str).apply(lambda x: int(x) * 1000)
addThreeZeros(df, ['Min_Salary', 'Max_Salary'])
```

```python
# as we extract useful information from 'Salary Estimate' we can drop it.
df.drop(columns=['Salary Estimate'], axis=1, inplace = True)
```

## 4.4 Resolving Noises from Company Name Variable <a id=4.4></a>

**As it is shown in below table , we can see that some company's name consists of "\n" and white space. It may happened during data integration or maybe during scraping the data from sources. We must handle and solve this issue.**

```python
df['Company Name'].value_counts().head(5)
```

```python
def companyName(co_name):
    """
    Cleans the company name by removing any non-alphabetic characters.

    Parameters:
    - co_name (str): The company name to be cleaned.

    Returns:
    - str: The cleaned company name containing only alphabetic characters.
    """
    co_name = ''.join(filter(str.isalpha, co_name))
    return co_name
```

```python
df['Company Name'] = df['Company Name'].apply(companyName).apply(pd.Series)
```

```python
df['Company Name'].value_counts().head(5)
```

```python
df.info()
```

## 4.5 Extract State Code from Location <a id=4.5></a>

```python
def extractLocation(loc):
    """
    Extracts the city from the location string.

    Parameters:
    - loc (str): The location string containing city and possibly other information.

    Returns:
    - str: The extracted city name.
    """
    return loc.split(',')[0].strip()
```

## 4.7 Revenue Categorization <a id=4.7></a>

```python
def group_categories(revenue):
    """
        Group revenue categories based on predefined criteria.

        Parameters:
        - revenue (str): The revenue value to be categorized.
        
        Returns:
        - str: Categorized revenue value.
    """
    
    if isinstance(revenue, str):
        if revenue == 'Unknown / Non-Applicable':
            return 'Unknown / Non-Applicable'
        elif 'billion' in revenue:
            return 'High Revenue'
        elif 'million' in revenue:
            return 'Medium Revenue'
        elif 'less than' in revenue:
            return 'Low Revenue'
        else:
            return revenue
```

```python
df['Revenue'] = df['Revenue'].apply(group_categories).apply(pd.Series)
```

```python
df['Revenue'].value_counts()
```
